```python
import  pandas as pd
import matplotlib.pyplot as plt
data={
    'Time':['Evening','Afternoon','Evening','Night','Morning','Evening','Afternoon','Night'],
    'Location':['North','South','East','West','Central','North','South','East'],
    'Item':['Biryani','Pizza','Dosa','Burger','Biryani','Biryani','Dosa','Pizza']
}
df=pd.DataFrame(data)
print("--- All Orders --- ")
print(df)
print("\n--- Peak Order Times ---")
print(df['Time'].value_counts())
print("\n---Most Popular Dishes ---")
popular=df['Item'].value_counts
print(popular)
biryani_orders=df[df['Item']=='Biryani']
print("\n--- Filtered Biryani Orders---")
print(biryani_orders)
plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
df['Time'].value_counts().plot(kind='bar',color='tomato')
plt.title('Peak Order Time')
plt.xlabel('Time')
plt.ylabel('Orders')
plt.subplot(1,2,2)
df['Item'].value_counts().plot(kind='bar',color='gold')
plt.title('Top Dishes')
plt.xlabel('food Item')
plt.ylabel('Orders')
plt.tight_layout()
plt.savefig('food_delivery_dashboard.png')
plt.show()
```
